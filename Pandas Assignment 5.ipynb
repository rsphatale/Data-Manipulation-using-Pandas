1. Use a for loop to calculate the number of customers that are getting the tech support and are male senior citizen.

```python
import pandas as pd
data = pd.read_csv("customer_churn.csv")
c = 0
for i in range(7043):
    if(data.loc[i,"gender"] == "Male" and data.loc[i,"SeniorCitizen"] ==1 and 
     data.loc[i,"TechSupport"]  == "Yes") :
        c = c + 1
print(c)
        
```
