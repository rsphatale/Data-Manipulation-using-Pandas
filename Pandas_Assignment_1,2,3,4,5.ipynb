1.  Write a function that takes start and end of a range returns a Pandas series object containing numbers within that range. In case the user does not pass start or end or both they should default to 1 and 10 respectively. 
eg. 

range_series() -> Should Return a pandas series from 1 to 10

range_series(5) -> Should Return a pandas series from 5 to 10 

range_series(5, 10) -> Should Return a pandas series from 5 to 10. 


```python
import pandas as pd
def range_series(a=1, b=10):
  return pd.Series(range(a,b+1))
range_series()
```

```python
range_series(5)
```

```python
range_series(5,10)
```

```python
cars.corr()
```

```python
import pandas as pd
df= pd.read_csv('/content/Customer_Churn.csv')
```

```python
x=0

for i in range(7043):
  if(df.loc[i,'gender']=='Male' and df.loc[i,'SeniorCitizen']==1 and df.loc[i, 'TechSupport']== 'Yes'):
    x=x+1
print(x)
```

```python
# using & instead of and
x=0
for i in range(7043):
  if((df.loc[i,'gender']=='Male') & (df.loc[i,'SeniorCitizen']==1) & (df.loc[i, 'TechSupport']== 'Yes')):
    x=x+1
print(x)
```
